```python
import VariationalAutoEncoder as VAE
from VariationalAutoEncoder import Trainer

trainer = Trainer()

vae = VAE.Model(
    VAE.Encoder(),
    VAE.Decoder(),
    epochs=10,
    optimizer='ADAM',
    learning_rate=0.01,
)

vae.train(trainer.train_set, batch_size=32, preview=True)
```

```python
vae.preview_annealing(50000)
```

```python
import VariationalAutoEncoder as VAE
tester = VAE.Tester(vae, trainer.test_set)
tester(1, 2)
```
